## Data cleaning

### Remove empty recipe entries

```python
ing.dropna(subset=['Directions'], inplace=True)
```

```python
assert len(ing[ing['Directions'].isnull()]) == 0
```

```python
len(ing)
```

### Remove ingredients related inconsistencies

```python
new_ing = pd.DataFrame(columns=['Ingredients', 'Directions'])

for i, r in ing.iterrows():
    ingredients = r.Ingredients.split(',')
    new_ingredients = []
    
    for e in ingredients:
        if e in r['Directions']:
            new_ingredients.append(e)
        else:
            aux = ' '.join(s for s in e.split() if s in r['Directions'])
            
            if aux != '':
                new_ingredients.append(aux)
                
    if new_ingredients:
        new_ing = new_ing.append({
            'Ingredients': ','.join(list(set(new_ingredients))),
            'Directions': r['Directions']
            },  ignore_index = True)
```

```python
len(new_ing)
```

### Remove duplicated recipes

```python
new_ing = new_ing.drop_duplicates(subset=['Ingredients', 'Directions'], keep=False)
```
